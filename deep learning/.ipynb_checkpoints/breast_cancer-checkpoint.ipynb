```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report
np.random.seed(10)
```

```python
X,y=load_breast_cancer(return_X_y=True)
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42)
```

```python
clf=MLPClassifier()
```

```python
clf.fit(X_train,y_train)
```

```python
pred=clf.predict(X_test)
```

```python
print(classification_report(y_test,pred))
```

### Weight initialization

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
num_features=X_train.shape[1]
l1_weights=np.random.random((num_features,16))
l1_bias=np.random.randn()
l2_weights=np.random.random((16,8))
l2_bias=np.random.randn()
l3_weights=np.random.random((8,1))
l3_bias=np.random.randn()
```

```python
def predict(l1_weights,l1_bias,l2_weights,l2_bias,l3_weights,l3_bias,input_data):
    input_layer=sigmoid(np.dot(input_data,l1_weights)+l1_bias)
    hidden1=sigmoid(np.dot(input_layer,l2_weights)+l2_bias)
    output=sigmoid(np.dot(hidden1,l3_weights)+l3_bias)
    return output
```

```python
def loss(y_hat,y):
    error=np.mean(-y*np.log(y_hat)+(1-y)*(np.log(1-y_hat)))
    return error
```

```python
def sigmoid_prime(x):
    return sigmoid(x)*(1-sigmoid(x))
```
